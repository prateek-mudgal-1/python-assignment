```python
# Key Features of Python
print("Python is easy to learn, dynamically typed, and supports multiple paradigms.")

# Predefined Keywords in Python
def greet(name):
    if name:
        return f"Hello, {name}"
    else:
        return "Hello, Guest"

print(greet("Alice"))

# Mutable vs Immutable Objects
# Mutable (List)
my_list = [1, 2, 3]
my_list.append(4)
print("Mutable List:", my_list)

# Immutable (Tuple)
my_tuple = (1, 2, 3)
try:
    my_tuple[0] = 10  # This will raise an error
except TypeError:
    print("Tuples are immutable")

# Different Types of Operators
# Arithmetic Operators
a, b = 10, 3
print("Addition:", a + b)
print("Exponentiation:", a ** b)

# Comparison Operators
print("10 > 5:", 10 > 5)

# Logical Operators
x, y = True, False
print("x AND y:", x and y)
print("NOT x:", not x)

# Bitwise Operators
print("Bitwise AND (5 & 3):", 5 & 3)
print("Bitwise OR (5 | 3):", 5 | 3)

# Assignment Operators
a = 10
a += 5
print("Updated a:", a)

# Type Casting
x = "100"
y = int(x)  # String to Integer
print("Type Casted Value:", y + 10)

# Conditional Statements
num = 10
if num > 0:
    print("Positive")
elif num < 0:
    print("Negative")
else:
    print("Zero")

# Loops in Python
# for Loop
for i in range(5):
    print("for loop iteration:", i)

# while Loop
count = 0
while count < 5:
    print("while loop iteration:", count)
    count += 1

# Nested Loops
for i in range(3):
    for j in range(2):
        print(f"Nested loop: i={i}, j={j}")
```
